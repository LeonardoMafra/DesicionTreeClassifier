```python
# Decision Tree Precision x Recall Curve
```

```python
#libraries

from sklearn    import datasets        as dt
from matplotlib import pyplot          as plt
from sklearn    import model_selection as ms
from sklearn    import tree            as tr
from sklearn    import metrics         as mt
from sklearn    import linear_model    as lm
```

```python
x,y = dt.make_classification(
    n_samples = 1000,
    n_classes = 2,
    random_state = 0
)
```

```python
# grafico para vizualização dos dados criados pelo make.classification

plt.scatter(x[:,0], x[:,1], c=y)
```

```python
#diviosao dos dados para treino teste e validação

x_train, x_test, y_train, y_test = ms.train_test_split(x,y , test_size = 0.2)
```

```python
# treinamento do modelo - definição - treinamento - performace
```

```python
# definicao do modelo
model = tr.DecisionTreeClassifier(
    max_depth = 2
)

# fit ou treinamento

model.fit(x_train, y_train)
```

```python
# performace

y_hat_test = model.predict_proba(x_test) # predicit_proba retorna a porcentagem de chace dos valores estarem em cada linha
yhat_test = y_hat_test[:, 1] # selecionando somente a segunda coluna dos valores retornados pelo predict_proba

```

```python
# precsion , recall and thereshrolds

precision, recall , thresholds = mt.precision_recall_curve(y_test, yhat_test)
```

```python
print(precision)
print(recall)
print(thresholds)
```

```python
#grasfico da curva precicion x recall

plt.plot(recall, precision, marker='.', label = model)
plt.xlabel('Recall')
plt.ylabel('Precision')
```

```python
x,y = dt.make_classification(
    n_samples = 500,
    n_classes = 2,
    n_redundant =2,
    random_state = 0
)
```

```python
x_train, x_test, y_train, y_test = ms.train_test_split(x,y , test_size = 0.2)
```

```python
# testando com alterações no modelo

# definicao do modelo
model = lm.LogisticRegression(
    solver = 'lbfgs'
)

# fit ou treinamento

model.fit(x_train, y_train)
```

```python
y_hat_test = model.predict_proba(x_test) # predicit_proba retorna a porcentagem de chace dos valores estarem em cada linha
yhat_test = y_hat_test[:, 1] # selecionando somente a segunda coluna dos valores retornados pelo predict_proba

```

```python
precision, recall , thresholds = mt.precision_recall_curve(y_test, yhat_test) # retornando precisao, recall e theresholds
```

```python
#grasfico da curva precicion x recall

plt.plot(recall, precision, marker='.', label = model)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
```

```python
# grafico para verificar as 3 dimensoes entre precision recall e theresholds


plt.plot(thresholds, precision[:-1], 'b--', label = 'Precision')
plt.plot(thresholds, recall[:-1], 'g-', label = 'Recall')
plt.xlabel('Thresholds')
plt.ylabel('Precision, Recall')
plt.legend()

plt.vlines(0.5, 0, 1, linestyle = '--', color = 'r')
plt.hlines(0.9, 0, 1, linestyle = '--', color = 'b')
plt.hlines(0.7, 0, 1, linestyle = '--', color = 'g')
```
