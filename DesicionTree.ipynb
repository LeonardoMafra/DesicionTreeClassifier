```python
#Imports

from sklearn    import datasets  as dt
from sklearn    import tree      as tr
from matplotlib import pyplot    as plt

import pydot
import cv2
```

```python
#Dataset

iris = dt.load_iris()

x = iris.data[:,2:]
y = iris.target
```

```python
#model training


# define

model_tree = tr.DecisionTreeClassifier(
   #max_depth = controle do tamanho maximo de crescimento ou quebras da arvore
   #min_samples_leaf = o numero minimo de amostras que a folha seguinte deve ter apos  a divisao
   #min_samples_split = o numero minimo de amostras para que a folha faça o proximo split
   #max_features = numero maximo de atributos avaliados por divisao

)

#training (fit)

model_tree.fit(x,y)

#predict
```

```python
# grafico para vizualizar os criterios que a desicion Tree usou para realizar a classificação - time de negocios
```

```python
# criando arquivo .dot
export_tree_data = tr.export_graphviz(
    model_tree,
    out_file = 'tree.dot',
    feature_names = iris.feature_names[2:],
    class_names = iris.target_names,
    rounded = True,
    filled = True
)
```

```python
# transformando o arquivo .dot em arquivo .png
(graph,) = pydot.graph_from_dot_file('tree.dot')
graph.write_png('tree.png')
```

```python
# lendo o arquivo png e plotando o grafico
img = cv2.imread('tree.png')
plt.figure(figsize=(20,20))
plt.imshow(img)
```
